```python
import requests
import pandas as pd

# Define your API key
API_KEY = "4aaf990a50fdeaf654ad6e6006d60fb6"

# Base URL for Scopus Search API
BASE_URL = "https://api.elsevier.com/content/search/scopus"

# Query parameters with COMPLETE view
params = {
    "query": "TITLE-ABS-KEY(Piper)",  # The search query
    "count": 25,  # Number of results per page
    "start": 0,   # Pagination: start at the first result
    "apikey": API_KEY,  # Your API key
    "view": "COMPLETE",  # Request detailed results including abstracts
}

# Function to retrieve and collect results
def fetch_scopus_results():
    results = []
    while True:
        response = requests.get(BASE_URL, params=params)
        if response.status_code != 200:
            print(f"Error: {response.status_code} - {response.text}")
            break

        data = response.json()
        entries = data.get("search-results", {}).get("entry", [])
        if not entries:  # Stop if no more entries
            break

        results.extend(entries)
        params["start"] += 25  # Move to the next set of results

    return results

# Convert Scopus results to a pandas DataFrame
def scopus_to_dataframe(results):
    # Extract relevant fields from each result
    records = []
    for item in results:
        record = {
            "Title": item.get("dc:title"),
            "DOI": item.get("prism:doi"),
            "Publication Name": item.get("prism:publicationName"),
            "Authors": item.get("dc:creator"),
            "Year": item.get("prism:coverDate", "").split("-")[0],  # Extract year
            "Scopus ID": item.get("dc:identifier"),
            "Abstract": item.get("dc:description"),  # Abstract retrieved directly
        }
        records.append(record)

    # Create a DataFrame
    return pd.DataFrame(records)

# Main script
if __name__ == "__main__":
    print("Fetching Scopus results...")
    results = fetch_scopus_results()
    print(f"Retrieved {len(results)} results.")

    if results:
        print("Converting to DataFrame...")
        df = scopus_to_dataframe(results)

        # Save to CSV (optional)
        df.to_csv("scopus_results_with_abstracts.csv", index=False)

        # Display the first few rows
        print(df.head())
    else:
        print("No results found.")
```

```python
import requests

# Define your API key
API_KEY = "4aaf990a50fdeaf654ad6e6006d60fb6"

# Base URL for Scopus Search API
BASE_URL = "https://api.elsevier.com/content/search/scopus"

# Query parameters
params = {
    "query": "TITLE-ABS-KEY(Piper)",  # The search query
    "count": 0,  # Don't return results, only metadata
    "apikey": API_KEY,
    "view": "STANDARD",  # Default view for basic metadata
}

# Fetch total number of results
def fetch_total_results():
    response = requests.get(BASE_URL, params=params)
    if response.status_code == 200:
        data = response.json()
        total_results = int(data.get("search-results", {}).get("opensearch:totalResults", 0))
        return total_results
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# Main script
if __name__ == "__main__":
    print("Fetching total number of results for query...")
    total_results = fetch_total_results()
    if total_results is not None:
        print(f"Total number of papers matching the query: {total_results}")
    else:
        print("Failed to retrieve the total number of results.")
```
